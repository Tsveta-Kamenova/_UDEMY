```python
import pandas as pd
import matplotlib.pyplot as plt
world = pd.read_csv('/kaggle/input/happiness-sales-airbnb-ether-salaries/2018-2019Happiness.csv')
airbnb = pd.read_csv('/kaggle/input/happiness-sales-airbnb-ether-salaries/airbnb.csv')
salaries = pd.read_csv('/kaggle/input/happiness-sales-airbnb-ether-salaries/salaries.csv')
sales = pd.read_csv('/kaggle/input/happiness-sales-airbnb-ether-salaries/web_sales.csv')
eth = pd.read_csv('/kaggle/input/happiness-sales-airbnb-ether-salaries/EthereumHistoricalData.csv', index_col = 0)
```

```python
avg_salary_experience = salaries.groupby('experience_level')\
                        ['salary_in_usd'].mean().reset_index()\
                        .sort_values(by='salary_in_usd', ascending = False)
```

```python
plt.bar(avg_salary_experience['experience_level'],avg_salary_experience['salary_in_usd'])
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.title('Average Salary against Experience')
```

```python
plt.figure(figsize = [4,3], dpi=120)
plt.plot(eth.index,eth['Price'], color = 'skyblue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Etherium Price chart')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize = [4,3], dpi=120)
plt.plot(eth.index,eth['High'], color = 'orange')
plt.xlabel('Date')
plt.ylabel('High')
plt.title('Etherium High chart')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize = [4,3], dpi=120)
plt.plot(eth.index,eth['Low'], color = 'red')
plt.xlabel('Date')
plt.ylabel('Low')
plt.title('Etherium Low chart')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize = [10,3], dpi=120)
plt.subplot(1,3,1)
plt.plot(eth.index,eth['Price'], color = 'skyblue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Etherium Price chart')
plt.xticks(rotation=45)

plt.subplot(1,3,2)
plt.plot(eth.index,eth['High'], color = 'orange')
plt.xlabel('Date')
plt.title('Etherium High Price chart')
plt.xticks(rotation=45)

plt.subplot(1,3,3)
plt.plot(eth.index,eth['Low'], color = 'red')
plt.xlabel('Date')
plt.title('Etherium Low Price chart')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
bike_sales = sales[sales['product_subcategory'].isin(['Road Bikes','Mountain Bikes','Touring Bikes'])]
```

```python
sales['product_subcategory'].unique()
```

```python
bike_sales.groupby('product_subcategory')['SalesAmount'].sum().reset_index()
```

```python
accessory_sales = sales[~sales['product_subcategory'].isin(['Road Bikes','Mountain Bikes','Touring Bikes'])]
```

```python
accessory_sales_group = accessory_sales.groupby('product_subcategory')\
                        ['SalesAmount'].sum().reset_index().sort_values(by = 'SalesAmount')
```

```python
bike_sales_group = bike_sales.groupby('product_subcategory')\
                        ['SalesAmount'].sum().reset_index().sort_values(by = 'SalesAmount', ascending = False)
```

```python
plt.pie(bike_sales_group['SalesAmount'], labels = bike_sales_group['product_subcategory'])
```

```python
plt.barh(accessory_sales_group['product_subcategory'], accessory_sales_group['SalesAmount'])
```

```python
sales['OrderDate'] = pd.to_datetime(sales['OrderDateKey'], format = '%Y%m%d')
sales['MonthYear'] = pd.to_datetime(sales['OrderDate'].dt.strftime('%Y-%m'))
sales
```

```python
monthly_sales = sales.groupby('MonthYear')['SalesAmount'].sum().reset_index()
daily_sales = sales.groupby('OrderDate')['SalesAmount'].sum().reset_index()
plt.plot(monthly_sales['MonthYear'],monthly_sales['SalesAmount'])
plt.show()
```

```python
plt.plot(daily_sales['OrderDate'],daily_sales['SalesAmount'])
plt.show()
```

```python
country_sales = sales.groupby('SalesTerritoryCountry')['SalesAmount']\
                .sum().reset_index().sort_values(by = 'SalesAmount', ascending = False)
```

```python
plt.bar(country_sales['SalesTerritoryCountry'],country_sales['SalesAmount'])
```

```python
plt.figure(figsize = (12,8))

plt.subplot(2,2,1)
#Line chart
plt.plot(monthly_sales['MonthYear'],monthly_sales['SalesAmount'], color = 'orange', marker = ".", linestyle = '--', linewidth = 2.5)
plt.xticks(rotation=45, fontsize = 8, color = 'black')
plt.title('Sales by Month', fontweight = 'bold')
plt.xlabel('Month-Year', fontsize = 10)
plt.ylabel('Sales', fontsize = 10)

plt.subplot(2,2,2)
#Horizontal Bar chart
plt.barh(accessory_sales_group['product_subcategory'], accessory_sales_group['SalesAmount'], color = 'green')
plt.title('Accessory Sales', fontweight = 'bold')

plt.subplot(2,2,3)
#Pie chart
plt.pie(bike_sales_group['SalesAmount'], labels = bike_sales_group['product_subcategory'], \
        autopct = '%1.1f%%',colors=['indianred', 'teal', 'lightskyblue'])
plt.title('Bike Sales', fontweight = 'bold')

plt.subplot(2,2,4)
#Bar chart
plt.bar(country_sales['SalesTerritoryCountry'],country_sales['SalesAmount'], color = 'steelblue')
plt.xticks(rotation=45, fontsize = 8, color = 'black')
plt.title('Country Sales', fontweight = 'bold')
plt.xlabel('Country', fontsize = 10)
plt.ylabel('Sales', fontsize = 10)
plt.tight_layout()
```
